```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)

```

```python
#Set the size of our figure
fig, ax = plt.subplots(figsize=(5, 3))
#Create a line plot with y1 and y2 against x
ax.plot(x,y1, y2)
```

```python
#Adding a legend
fig, ax = plt.subplots(figsize=(5, 3))
#Create a line plot with y1 and y2 against x
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
#Location of the legend
ax.legend (loc ='lower right',
           #Add a frame around the legend
           frameon = True,
           #Set the font size of the legend text
           fontsize = 12,
           #Set the number of columns in the legend
           ncol = 2,
           #Add a shadow to the legend
           shadow = True)
#Show the figure
fig.show()
```

```python
#Move the legend outside of the plot area
fig, ax = plt.subplots(figsize=(5, 3))
#Create a line plot with y1 and y2 against x
ax.plot(x,y1, label="Person 1")
ax.plot(x,y2, label="Person 2")
#Location of the legend
ax.legend (loc ='upper left',
           bbox_to_anchor =(1, 1))
fig.show()
```

```python
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"]) 
y2 = np.array([110, 180, 240, 99, 220])
```

```python
font2 = {'family':'serif', 'color':'hotpink'}
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1,
                               figsize=(7,3))

ax1.scatter(x1,
            y1, 
            marker = '*',
            color = 'pink')
ax2.plot(x2,
         y2, 
         color = "blue")

ax1.set_title('Scatter Plot', fontsize = 14, fontdict = font2)
ax1.set_ylabel('Numbers', fontsize = 14, fontdict = font1)
ax1.set_xlabel('More numbers', fontsize = 14, fontdict = font1)
ax2.set_title('Line Plot', fontsize = 14, fontdict = font2)
ax2.set_ylabel('Random Values', fontsize = 14, fontdict = font1)
ax2.set_xlabel('Random names', fontsize = 14, fontdict = font1)
plt.xticks(rotation = 45, ha = 'right', fontsize = 12, font = 'serif')  

```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4),
                                    layout = "constrained")

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2) 
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Lebel 2: 2 Fast 2 Furious', fontsize = 18)
```

```python
#Multiple viz on one axes object
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"]) 
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

#define our figure and axes
fig, ax = plt.subplots(figsize=(7, 3))
#now call both bar and plot elements to the sames axes (ax)
ax.bar(x, y1, 
       color = "indigo")
ax.plot(x, y2,
            color = "red")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x,y2,color = "red")
ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor = "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery = 2)
```

```python
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images

```

```python
response = requests.get('https://static.wikia.nocookie.net/the-ultimate-good/images/3/3d/Luffy.png/revision/latest?cb=20230417200546')
image_file = BytesIO(response.content)
image = Image.open(image_file)     
image.show()

```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
        
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
                         0.1, # y coordinate (ON FIGURE, NOT AXES)
                         0.15, # image width
                         0.35])  # image height

ax_image.imshow(image)
ax_image.axis('off')  # Hide axes for the image
```
